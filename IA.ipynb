```
quantidade = pd.value_counts(dataset['TP_COMPARECIMENTO'].values, sort= True)
```

```
print(quantidade)
```

### **Mesárias e mesários por tipo de convocação**

```
# Setting labels for items in Chart
Employee = ['Voluntário', 'Não Voluntário']
  
# Setting size in Chart based on 
# given values
Salary = [quantidade['Ausências'], quantidade['Comparecimento']]
  
# colors
colors = ['#FF0000', '#0000FF']
  
# Pie Chart
plt.pie(Salary, colors=colors, labels=Employee)
  
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
  
# Adding Title of chart
plt.title('Mesárias e mesários por tipo de convocação')
  
# Displaing Chart
plt.show()
```

```
plt.figure(figsize = (15,10))
sns.set(style = 'darkgrid')  
ax = sns.countplot(y=dataset["DS_FAIXA_ETARIA"], order = dataset["DS_FAIXA_ETARIA"].value_counts().index) 
ax.set(ylabel='quantidade', xlabel='faixa etária')
```
